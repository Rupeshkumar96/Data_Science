```python
## Find Total Quantity
Total_Quan=df['quantity'].sum()
print(Total_Quan)
```

```python
## Average Sales Transaction
avg_sale=Total_sales/Total_trans
print("Average sales per transaction",avg_sale)
```

```python
df['Update_date']=pd.to_datetime(df['sale_date'])
```
